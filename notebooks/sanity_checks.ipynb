```python
# años detectados por estación
import pandas as pd

def years_for(station):
    ys = []
    for p in paths:
        if f"isd_{station}_" in p:
            m = re.search(rf"isd_{station}_(\d{{4}})\.parquet$", p)
            if m: ys.append(int(m.group(1)))
    return sorted(set(ys))

print("GCXO years:", years_for("GCXO")[:5], "...", years_for("GCXO")[-5:])
print("GCTS years:", years_for("GCTS")[:5], "...", years_for("GCTS")[-5:])
```
