## 5 claims válidos (sí puedes afirmar)

1. **Asociación lagged:**
   “En Gran Canaria (2018+), las semanas con CAP dust yellow+ se asocian con un aumento de muertes **1–2 semanas después** (especialmente lag2), tras ajustar por semana del año y año.”

2. **El efecto no es contemporáneo:**
   “No observamos evidencia clara de un efecto contemporáneo (lag0) comparable al de lag1–lag2.”

3. **Robustez a incertidumbre por episodios:**
   “La diferencia de muertes desestacionalizadas en lag2 se mantiene positiva bajo bootstrap por episodios (IC95% > 0).”

4. **Placebo temporal pasa:**
   “Un lead test (CAP futuro) no predice muertes actuales, lo que sugiere que el patrón lagged no es un artefacto temporal obvio.”

5. **PM10 continuo no muestra señal lineal en ese setup:**
   “En este modelo (con controles estacionales), PM10 como variable continua no mostró asociación lineal significativa con muertes (ni contemporánea ni lag1).”

## 5 claims NO válidos (no los puedes afirmar con lo hecho)

1. **Causalidad directa:**
   ❌ “La calima/alerta CAP **causa** +15 muertes.”
   (Tu diseño no prueba causalidad; confusores no medidos pueden explicar parte.)

2. **Mecanismo médico específico:**
   ❌ “Las muertes aumentan por inflamación respiratoria causada por polvo sahariano.”
   (No mediste mecanismos, diagnósticos, ni causas de muerte.)

3. **Generalización fuera de muestra:**
   ❌ “Esto ocurrirá igual en todas las islas / en cualquier año futuro.”
   (Solo probaste GC y un periodo; faltan réplicas y validación externa.)

4. **CAP captura “toda” la exposición real:**
   ❌ “CAP yellow+ es una medida completa y precisa de exposición individual al polvo.”
   (Es un proxy/alerta; no mide exposición personal ni distribución espacial.)

5. **Exclusión de alternativas:**
   ❌ “No puede ser gripe/meteorología/otros contaminantes; tiene que ser polvo.”
   (Aún no has controlado esos confusores explícitamente.)

### Plantilla rápida “qué NO puedo afirmar”

* “No puedo inferir causalidad a partir de este análisis observacional.”
* “No puedo atribuir el efecto a un mecanismo específico o a un contaminante único.”
* “No puedo generalizar sin replicación en otras islas/periodos.”
* “No puedo descartar confusión residual (virus, otros contaminantes, meteorología detallada).”
* “No puedo asegurar que CAP represente exposición individual.”


#### “Observed association, not causation.”
#### “Result is conditional on this dataset, this definition, and these controls.”

1. **“No significativo” ≠ “no efecto”**: solo significa “con este modelo/datos no puedo distinguirlo de 0”.
2. PM10 puede salir no significativo por **confusión estacional**, **no linealidad/umbrales**, **timing (lags)**, **colinealidad** y **outliers extremos**.
3. Para ser justo con PM10, las siguientes pruebas naturales serían: **flag p95**, **log(PM10)**, o **lags 0–2 juntos**.
