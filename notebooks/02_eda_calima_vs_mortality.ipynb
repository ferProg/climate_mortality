```python
import pandas as pd
from pathlib import Path


# Find project root by walking up until we see the "data" folder
root = Path.cwd()
while root != root.parent and not (root / "data").exists():
    root = root.parent

print("Project root:", root)

fp = root / "data" / "processed" / "dust_days_tfe_tfs_tfn_2016_2024.parquet"
print("Loading:", fp)
df = pd.read_parquet(fp)
print(df.head())

# --- quick visibility diagnostics ---
for col in ["vis_m_gcxo", "vis_m_gcts"]:
    print(f"\n{col}")
    print(df[col].describe())
    print("unique small values sample:", sorted(df[col].dropna().unique())[:20])
    print("quantiles:", df[col].quantile([0.01, 0.05, 0.5, 0.95, 0.99]).to_dict())

print("\nMedian vis by station when RH < 70%:")
print("GCTS:", df.loc[df["rh_pct_gcts"] < 70, "vis_m_gcts"].median())
print("GCXO:", df.loc[df["rh_pct_gcxo"] < 70, "vis_m_gcxo"].median())
```
