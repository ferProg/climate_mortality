#Parche para missing columns for weeks in quality_air

```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().parent  # notebooks/

island = "gomera"   # cambia: tenerife / lanzarote / fuerteventura / ...
code = "gom"               # cambia: tfe / lzt / ftv / ...

in_fp = PROJECT_ROOT / "data" / "processed" / island / "air_quality" / f"air_quality_weekly_{code}_2016_2024.parquet"
out_fp = PROJECT_ROOT / "data" / "processed" / island / "air_quality" / f"air_quality_weekly_{code}_2016_2024_calendarized.parquet"

df = pd.read_parquet(in_fp).copy()

full_weeks = pd.date_range(df["week_start"].min(), df["week_start"].max(), freq="W-MON")
spine = pd.DataFrame({"week_start": full_weeks})

df2 = spine.merge(df, on="week_start", how="left").sort_values("week_start").reset_index(drop=True)
df2["n_days"] = df2["n_days"].fillna(0).astype(int)
df2["coverage"] = df2["coverage"].fillna(0.0)

df2.to_parquet(out_fp, index=False)

print("Saved:", out_fp)
print("rows before/after:", len(df), "->", len(df2))
print("dup weeks:", int(df2.duplicated(['week_start']).sum()))
```

```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().parent  # notebooks/
in_fp = PROJECT_ROOT / "data" / "processed" / "fuerteventura" / "air_quality" / "air_quality_weekly_ftv_2016_2024.parquet"
out_fp = PROJECT_ROOT / "data" / "processed" / "fuerteventura" / "air_quality" / "air_quality_weekly_ftv_2016_2024_calendarized.parquet"

df = pd.read_parquet(in_fp).copy()

# Build full weekly spine (Mondays) between min/max
full_weeks = pd.date_range(df["week_start"].min(), df["week_start"].max(), freq="W-MON")
full = pd.DataFrame({"week_start": full_weeks})

# Left merge to fill missing weeks
df2 = full.merge(df, on="week_start", how="left").sort_values("week_start").reset_index(drop=True)

# For missing weeks: set n_days/coverage = 0 (keep pollutant stats NaN)
df2["n_days"] = df2["n_days"].fillna(0).astype(int)
df2["coverage"] = df2["coverage"].fillna(0.0)

df2.to_parquet(out_fp, index=False)

print("Saved:", out_fp, "rows:", len(df2))
print("missing weeks after:", int(pd.date_range(df2["week_start"].min(), df2["week_start"].max(), freq="W-MON")
                                  .difference(pd.DatetimeIndex(df2["week_start"])).shape[0]))
print("coverage==0:", int((df2["coverage"] == 0).sum()))
```

```python
expected = pd.date_range(df2["week_start"].min(), df2["week_start"].max(), freq="W-MON")
missing = expected.difference(pd.DatetimeIndex(df2["week_start"].unique()))
print("missing weeks:", len(missing))
print("coverage==0:", int((df2["coverage"]==0).sum()))
```
