```python
import pandas as pd

fp = r"C:\dev\projects\heat_mortality_analysis\data\interim\tenerife\visibility\daily_island\visibility_daily_tfe_2016-01-01_2024-12-31.parquet"
df = pd.read_parquet(fp)

print("shape:", df.shape)
print("\ncolumns:")
print(df.columns.tolist())

print("\nhead:")
print(df.head(10))

print("\ndtypes:")
print(df.dtypes)

print("\nmissing pct:")
print(df.isna().mean().sort_values(ascending=False))

vis_cols = [c for c in df.columns if "vis" in c.lower()]
rh_cols = [c for c in df.columns if "rh" in c.lower()]
print("\nvis_cols:", vis_cols)
print("rh_cols:", rh_cols)

for c in vis_cols + rh_cols:
    print(f"\n=== {c} ===")
    print(df[c].describe(percentiles=[0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99]))
    print("\nvalue_counts head:")
    print(df[c].value_counts(dropna=False).head(20))
    if pd.api.types.is_numeric_dtype(df[c]):
        print("9999 count:", int((df[c] == 9999).sum()))
        print(">=9999 count:", int((df[c] >= 9999).sum()))
        print("==50000 count:", int((df[c] == 50000).sum()))
        if "date" in df.columns:
            date_col = "date"
        elif "obs_date" in df.columns:
            date_col = "obs_date"
        else:
            date_col = None
        
        if date_col:
            print("\nnon-null vis by year:")
            tmp = df.copy()
            tmp[date_col] = pd.to_datetime(tmp[date_col], errors="coerce")
            for c in vis_cols:
                print(f"\n{c}")
                print(tmp.groupby(tmp[date_col].dt.year)[c].apply(lambda s: s.notna().sum()))
```
